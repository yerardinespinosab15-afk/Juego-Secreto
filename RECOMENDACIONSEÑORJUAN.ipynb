### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Sumamos el precio de todos los productos vendidos en cada tienda
facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"Facturación Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Usamos value_counts para contar cuántas veces aparece cada categoría
print("Categorías más populares Tienda 1:")
print(tienda['Categoría del Producto'].value_counts().head(3))

print("\nCategorías más populares Tienda 2:")
print(tienda2['Categoría del Producto'].value_counts().head(3))

# Repetir lo mismo para tienda3 y tienda4
```

# 3. Calificación promedio de la tienda


```python
# Calculamos el promedio de la columna 'Calificación'
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {promedio_t1:.2f}")
print(f"Calificación promedio Tienda 2: {promedio_t2:.2f}")
print(f"Calificación promedio Tienda 3: {promedio_t3:.2f}")
print(f"Calificación promedio Tienda 4: {promedio_t4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Producto más vendido (el que más se repite)
mas_vendido_t1 = tienda['Producto'].value_counts().idxmax()
# Producto menos vendido (el que menos se repite)
menos_vendido_t1 = tienda['Producto'].value_counts().idxmin()

print(f"Tienda 1 - Más vendido: {mas_vendido_t1} | Menos vendido: {menos_vendido_t1}")
# Repetir para las demás tiendas
```

# 5. Envío promedio por tienda

```python
# Promedio de la columna 'Costo de envío'
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo envío promedio Tienda 1: ${envio_t1:,.2f}")
# Imprimir los demás resultados
```

```python
# 1. Preparar los datos (Asegúrate de haber corrido la celda de importación primero)
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
resultados = []

for i, df in enumerate(tiendas):
    dict_tienda = {
        "Tienda": nombres[i],
        "Facturación Total": df['Precio'].sum(),
        "Calificación Promedio": df['Calificación'].mean(),
        "Costo Envío Promedio": df['Costo de envío'].mean(),
        "Producto Estrella": df['Producto'].value_counts().idxmax(),
        "Categoría Top": df['Categoría del Producto'].value_counts().idxmax()
    }
    resultados.append(dict_tienda)

# 2. Crear tabla comparativa
df_comparativo = pd.DataFrame(resultados)
print("--- RESUMEN PARA TOMA DE DECISIONES ---")
print(df_comparativo)

# 3. Identificar la tienda con menor desempeño
tienda_a_vender = df_comparativo.loc[df_comparativo['Facturación Total'].idxmin()]
print(f"\nRECOMENDACIÓN: El señor Juan debería considerar vender la {tienda_a_vender['Tienda']}.")
```
